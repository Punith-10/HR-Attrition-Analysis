```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/Copy of People Charm_data.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
df.duplicated().sum()
df
```

```python
df['left'].sum()
```

```python
total_employees=df.shape[0]
employees_left=df['left'].sum()
attrition_rate = (employees_left / total_employees) * 100
print("Overall Attrition Rate:", round(attrition_rate, 2), "%")
plt.figure()
plt.pie(df['left'].value_counts(),labels=df['left'].value_counts().index,autopct='%1.1f%%',explode=(0,0.2),shadow=True)
plt.title('Attrition Rate')
plt.show()
```

```python
df.corr(numeric_only=True)['left'].sort_values(ascending=False)
```

```python
numeric_df = df.select_dtypes(include=['int64','float64'])
corr = numeric_df.corr()
plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Between Employee Factors and Attrition")
plt.show()
```

```python
plt.figure(figsize=(6,4))
ax = sns.barplot(x='dept', y='left', data=df)

plt.xticks(rotation=45)
plt.xlabel("Department")
plt.ylabel("Attrition Rate")
plt.title("Attrition Rate by Department")

for container in ax.containers:
    ax.bar_label(container, fmt='%.2f')

plt.show()
```

```python
df.groupby('dept')['left'].mean().sort_values(ascending=False) * 100
```

```python
sns.barplot(data=df,x='salary',y='left')
plt.title('Attrition Rate by Salary')
plt.xlabel('Salary')
plt.ylabel('Attrition Rate')
plt.show()
```
